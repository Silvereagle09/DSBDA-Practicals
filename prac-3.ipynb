```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("Titanic-Dataset.csv")
```

```python
grouped_stats = data.groupby("Sex")[["Age", "Fare"]].agg(
    ["mean", "median", "min", "max", "std"]
)

print(grouped_stats)
```

```python
print("\nPercentiles (Fare):")
print(data.groupby("Sex")["Fare"].quantile([0.25, 0.5, 0.75]))
```

```python
print("\nVariance (Fare):")
print(data.groupby("Sex")["Fare"].var())
```

```python
male_fare = data[data["Sex"] == "male"]["Fare"].dropna().tolist()
female_fare = data[data["Sex"] == "female"]["Fare"].dropna().tolist()
```

```python
sns.boxplot(x="Sex", y="Fare", data=data)
plt.title("Fare Distribution by Gender")
plt.show()
```
