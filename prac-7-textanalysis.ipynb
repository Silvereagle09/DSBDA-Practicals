```python
import nltk

# Tokenization
from nltk.tokenize import word_tokenize

# Stopwords
from nltk.corpus import stopwords

# POS Tagging
from nltk import pos_tag

# Stemming
from nltk.stem import PorterStemmer

# Lemmatization
from nltk.stem import WordNetLemmatizer

# Frequency
from collections import Counter

# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')  # needed in newer versions
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')
```

```python
text = """Text analytics is the process of analyzing textual data to extract meaningful insights.
It involves techniques such as tokenization, stemming, and lemmatization."""
```

```python
tokens = word_tokenize(text)
print("Tokens:\n", tokens)
```

```python
pos_tags = pos_tag(tokens)
print("\nPOS Tags:\n", pos_tags)
```

```python
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("\nAfter Stop Words Removal:\n", filtered_tokens)
```

```python
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_tokens]

print("\nStemmed Words:\n", stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("\nLemmatized Words:\n", lemmatized_words)
```

```python
tf = Counter(filtered_tokens)

print("\nTerm Frequency:\n", tf)
```

```python
documents = [text]

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

print("\nFeature Names:\n", vectorizer.get_feature_names_out())
print("\nTF-IDF Matrix:\n", tfidf_matrix.toarray())
```
